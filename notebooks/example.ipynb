```python
from sub_tsmd import SubTSMD, plot_motif_sets, load
```

```python
import seaborn as sns
sns.set_theme()
```

```python
motif_discovery = SubTSMD(
    l_min=40,
    l_max=60,
    rho=0.65,
    warping=False
)
```

## Subspace motif sets

```python
X, _ = load('../data/synthetic/demonstration/subspace.pkl')
subspace_motifs = motif_discovery.apply(X)
fig = plot_motif_sets(X, subspace_motifs, figsize=(20, 8.5))
for ax in fig.axes:    
    ax.set_yticklabels([])
    ax.set_xticklabels([])
fig.tight_layout()
fig.savefig('example.png')
```

## Full-dimensional motif sets

```python
X, y = load('../data/synthetic/demonstration/full-dimensional.pkl')
subspace_motifs = motif_discovery.apply(X)
plot_motif_sets(X, subspace_motifs);
```
