```python
import numpy as np
from sub_tsmd import SubTSMD, load, plot_motif_sets_marking
```

```python
import seaborn as sns
sns.set_theme()
```

```python
X, y = load('../data/synthetic/demonstration/subspace.pkl')
plot_motif_sets_marking(X, y);
```

```python
sub_tsmd = SubTSMD(45, 55)
y_pred = sub_tsmd.apply(X)
plot_motif_sets_marking(X, y_pred);
```

### Masks

```python
mask = np.zeros(shape=X.shape[0], dtype=bool)
mask[X.shape[0]//2:] = True
y_pred = sub_tsmd.apply(X, start_mask=mask, end_mask=mask)
plot_motif_sets_marking(X, y_pred);
```

### Inclusion constraint

```python
y_pred = sub_tsmd.apply(X, inclusion_constraint=[2])
plot_motif_sets_marking(X, y_pred);
```

```python
y_pred = sub_tsmd.apply(X, inclusion_constraint=[1, 2])
plot_motif_sets_marking(X, y_pred);
```

### Exclusion constraint

```python
y_pred = sub_tsmd.apply(X, exclusion_constraint=[1, 2])
plot_motif_sets_marking(X, y_pred);
```

### Size constraint

```python
y_pred = sub_tsmd.apply(X, size_constraint=3)
plot_motif_sets_marking(X, y_pred);
```

### Co-occurrence constraint

```python
y_pred = sub_tsmd.apply(X, co_occurrence_constraint=[[4, 5]])
plot_motif_sets_marking(X, y_pred);
```

### Multiple constraints

```python
y_pred = sub_tsmd.apply(X, exclusion_constraint=[1], size_constraint=2, co_occurrence_constraint=[[4, 5]])
plot_motif_sets_marking(X, y_pred);
```
